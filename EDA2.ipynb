**Encoding Techniques**

```python
categorical_cols=df.select_dtypes(include='object').columns
categorical_cols
```

```python
for col in categorical_cols:
    b=df[col].unique()
    print(f'{col} had {b}','\n')
```

```python
onehot=OneHotEncoder()
label=LabelEncoder()
```

```python
dummies=pd.get_dummies(df[['sex','income']],dtype=int)
dummies.head()
```

**Label Encoder**

```python
df['education'] = label.fit_transform(df['education'])
df['workclass'] = label.fit_transform(df['workclass'])
df['marital_status'] = label.fit_transform(df['marital_status'])
df['occupation'] = label.fit_transform(df['occupation'])
df['relationship'] = label.fit_transform(df['relationship'])
df['race'] = label.fit_transform(df['race'])
df['native_country'] = label.fit_transform(df['native_country'])
```

```python
df.drop(df[['sex','income']],axis=1,inplace=True)
```

```python
df=df.join(dummies,how='right')
```

```python
df
```

```python
df.info()
```

**3. Feature Engineering**

```python
df['capital_gain-minus-loss']=df['capital_gain'] - df['capital_loss']
df['age_squared']=df['age']**2
```

```python
df
```

```python
df['capital_gain_log']=np.log(df['capital_gain'])
```

```python
df.head()
```

**4.Feature selection**

```python
ift=IsolationForest(n_estimators=10,contamination=0.1,random_state=42)
```

```python
ift.fit(df)
```

```python
outliers=pd.DataFrame(ift.predict(df))
```

```python
outliers.columns = ['outliers']
```

```python
df=df.join(outliers,how='right')
```

```python
df.head()
```
